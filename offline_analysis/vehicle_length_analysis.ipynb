# Calculating the Effective Vehicle Length in Tuscaloosa

Used Google Maps images to count cars and measure length

```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

class Data:
    def __init__(self):
        self.McFarland_15th = pd.DataFrame()
        self.McFarland_Skyland = pd.DataFrame()
        self.Lurleen = pd.DataFrame()
        self.Total = pd.DataFrame()
        
data = Data()
```

### 15th & McFarland

NB

```python
length = [0]*6
vehicle_num = [0]*6

# Lane 6
length[5] = 45.20 #ft
vehicle_num[5] = 2

# Lane 5
length[4] = 92.68 #ft
vehicle_num[4] = 4

# Lane 4
length[3] = 269.34 #ft
vehicle_num[3] = 10

# Lane 3
length[2] = 134.78 #ft
vehicle_num[2] = 5

# Lane 2
length[1] = 71.23 #ft
vehicle_num[1] = 3

# Lane 1
length[0] = 127.45 #ft
vehicle_num[0] = 5

# for i, val in enumerate(zip(length, vehicle_num)):
#     print(f"Lane {i+1} has an effective vehicle length of {val[0]/val[1]} ft")

# print(f"The total average is {sum(length)/sum(vehicle_num)} ft")

data.McFarland_15th['length'] = length
data.McFarland_15th['vehicle_num'] = vehicle_num
data.McFarland_15th['effective_length'] = data.McFarland_15th['length'] / data.McFarland_15th['vehicle_num']

data.McFarland_15th['length'].sum() / data.McFarland_15th['vehicle_num'].sum()
data.McFarland_15th.head()
data.McFarland_15th.describe()

data.McFarland_15th
```
